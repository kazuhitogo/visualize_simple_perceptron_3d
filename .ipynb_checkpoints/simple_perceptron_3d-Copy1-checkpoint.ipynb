```python
%matplotlib nbagg
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
```

```python
stable_rand = np.random.RandomState(4545)
```

```python
# w0 * x0 + w1 * x1 + w2 * x2 + b = 0
# x2 = -(w0/w2)*x0 - (w1/w2)*x1 - b/w2 = 0
N = 10000

w0 = 5
w1 = -3
w2 = 2
b = 1

# ラベル1
x0_rand_posi = stable_rand.rand(int(N/2))
x1_rand_posi = stable_rand.rand(int(N/2))
x2_rand_posi = -(w0/w2)*x0_rand_posi - (w1/w2)*x1_rand_posi -b/w2 + stable_rand.rand(int(N/2))

# ラベル2
x0_rand_nega = stable_rand.rand(int(N/2))
x1_rand_nega = stable_rand.rand(int(N/2))
x2_rand_nega = -(w0/w2)*x0_rand_nega - (w1/w2)*x1_rand_nega -b/w2 - stable_rand.rand(int(N/2))
```

```python
train_x = np.array([np.append(x0_rand_posi,x0_rand_nega,axis=0),np.append(x1_rand_posi,x1_rand_nega,axis=0),np.append(x2_rand_posi,x2_rand_nega,axis=0)])
train_y = np.append(np.ones(int(N/2)),np.zeros(int(N/2)))
```

```python
train_x.T.shape
```

```python
np.savetxt('train_x.csv',train_x,delimiter=',')
np.savetxt('train_y.csv',train_y,delimiter=',')
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("x2")
ax.set_xlim3d(0, 1)
ax.set_ylim3d(0, 1)
ax.set_zlim3d(-3, 1)
ax.plot(train_x[0,:int(N/2)],train_x[1,:int(N/2)],train_x[2,:int(N/2)],"o",ms=4, mew=0.5,color="#00ff00")
ax.plot(train_x[0,int(N/2):],train_x[1,int(N/2):],train_x[2,int(N/2):],"o",ms=4, mew=0.5,color="#ff0000")
ax.view_init(19, 97)
plt.show()
```

```python
w = np.ones(3)
b = np.ones(1)
```

```python
def step(x):
    return 1 * (x > 0)
def predict(w,x,b):
    return step(np.dot(w, x) + b)    
```

```python
%%time
eps = .01
while True:
    fin_flag = True
    for i in range(len(train_y)):
        if i % 2 == 0:
            idx = int(i/2)
        else:
            idx = int((i-1)/2) + 500
        delta_w = eps * (train_y[idx] - predict(w,train_x[:,idx],b)) * train_x[:,idx]
        delta_b = eps * (train_y[idx] - predict(w,train_x[:,idx],b))    
        w += delta_w
        b += delta_b
        # 一回でも誤り判定があればTrueになる
        fin_flag *= all(delta_w == 0) * (delta_b == 0)
    if fin_flag:
        break
        
```

```python
w = np.ones(3)
b = np.ones(1)
```

```python
def step(x):
    return 1 * (x > 0)
def dot(a,b):
    ab = 0
    for ai,bi in zip(a,b):
        ab += ai*bi
    return ab
def predict(x):
    return step(dot(w, x) + b)    
```

```python
%%time
eps = .01
while True:
    fin_flag = True
    for i in range(len(train_y)):
        if i % 2 == 0:
            idx = int(i/2)
        else:
            idx = int((i-1)/2) + 500
        delta_w = eps * (train_y[idx] - predict(train_x[:,idx])) * train_x[:,idx]
        delta_b = eps * (train_y[idx] - predict(train_x[:,idx]))    
        w += delta_w
        b += delta_b
        # 一回でも誤り判定があればTrueになる
        fin_flag *= all(delta_w == 0) * (delta_b == 0)
    if fin_flag:
        break
```
